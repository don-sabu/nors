```python
import pandas as pd
```

```python
train_data = {
    'order_id': [2232, 2232, 2232, 3432, 3432, 3432, 3432],
    'product_id': [43, 453, 347, 545, 976, 765, 533]
}
train_df = pd.DataFrame(train_data)
train_df
```

```python
category_data = {
    'product_id': [43, 453, 347, 545, 976, 765, 533],
    'category_id': [21, 36, 21, 21, 45, 36, 21]
}
category_df = pd.DataFrame(category_data)
category_df
```

```python
merged_df = pd.merge(train_df, category_df, on='product_id')
pivot_df = merged_df.pivot_table(index='order_id', columns='category_id', aggfunc='size', fill_value=0)
pivot_df
```

```python
pivot_df.columns = [str(col) for col in pivot_df.columns]
pivot_df.reset_index(inplace=True)
pivot_df
```

```python
path = 'minimized_data/804/'
```

```python
train_df = pd.read_csv(path+'train.csv')
train_df
```

```python
unique_id_list = pd.read_csv('prodid_to_uniqueid_mapped.csv')
unique_id_list.rename(columns={'id': 'product_id'}, inplace=True)
unique_id_list
```

```python
merged_df = pd.merge(train_df, unique_id_list, on='product_id')
pivot_df = merged_df.pivot_table(index='order_id', columns='unique_id', aggfunc='size', fill_value=0)
pivot_df
```

```python
all_unique_ids = unique_id_list['unique_id'].unique()

zeros_df = pd.DataFrame(0, index=pivot_df.index, columns=[uid for uid in all_unique_ids if uid not in pivot_df.columns])

pivot_df = pd.concat([pivot_df, zeros_df], axis=1)

# Sort columns by category ID
pivot_df = pivot_df.reindex(sorted(pivot_df.columns), axis=1)
pivot_df
```
