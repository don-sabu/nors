### Modify the dataset to only contain unique product id

```python
import pandas as pd
import os
```

```python
df = pd.read_csv('data/804/train.csv')
df
```

```python
df_filtered = df[df['order_number'] != 99999999]
df_filtered
```

```python
users = os.listdir("data/")
for user in users:
    train = pd.read_csv("data/" + user + "/train.csv")
    filtered = train[train['order_number'] != 99999999]
    os.makedirs("minimized_data/" + user)
    filtered.to_csv('minimized_data/' + user + '/train.csv', index=False)
    print(user)
```
