```python
import numpy as np
import pandas as pd

# Function to generate the dataset
def generate_dataset(num_orders, num_products):
    # Set the random seed for reproducibility
    np.random.seed(42)

    # Generate non-sequential five-digit order IDs
    order_ids = np.random.choice(range(10000, 100000), num_orders, replace=False)

    # Generate random binary data for the orders
    order_data = np.random.randint(0, 2, size=(num_orders, num_products))

    # Create a DataFrame for the orders
    orders_df = pd.DataFrame(order_data, columns=[str(i) for i in range(1, num_products + 1)])
    orders_df.insert(0, 'order_id', order_ids)

    return orders_df
```

```python
orders_dataset = generate_dataset(20000, 20)
orders_dataset.to_csv('sample_data/order_list_20k.csv', index=False)
```

```python
def generate_patterned_dataset(num_orders, num_products, frequencies):
    np.random.seed(42)
    order_ids = np.random.choice(range(100000, 1000000), num_orders, replace=False)
    data = np.zeros((num_orders, num_products), dtype=int)

    for idx, freq in enumerate(frequencies):
        # Calculate the probability of purchase based on the frequency
        probability = 1 / freq
        purchases = np.random.binomial(1, p=probability, size=num_orders)
        data[:, idx] = purchases

    # Create the DataFrame
    orders_df = pd.DataFrame(data, columns=[str(i) for i in range(1, num_products + 1)])
    orders_df.insert(0, 'order_id', order_ids)
    return orders_df
```
